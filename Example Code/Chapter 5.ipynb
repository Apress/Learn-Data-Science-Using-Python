## Listing 5.2 While Loop

```python
input_no = int(input('Enter your number (0 to exit): '))
while input_no != 0:

    if input_no%2==0:
        print("your input is even")
    else:
        print("your input is odd")
    input_no = int(input('Enter your number (0 to exit): '))
else:
    print("Bye and Have a great day!")
```

## Listing 5.3 For Loop over a sequence of numbers

```python
sum=0
for num in range(1,10):
    if num%2!=0:
        sum=sum+num
        print('num=',num)
        print('sum=',sum)
```

## Listing 5.4 For Loop over a string

```python
for letter in 'Python'[::-1]:
    print(letter)
```

## Listing 5.5 For Loop over a string using range()

```python
p='Python'
for i in range(len(p),0,-1):
    print(p[i-1])
```

## Listing 5.6 For Loop over a list

```python
my_list=['I','Love',"Python"]
for c in my_list:
    print(c,end="^")
```

## Listing 5.7 For Loop over a DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


students_df = pd.read_excel('../Datasets/Chapter 5/missing_null.xlsx')

print(students_df,'\n\n\n')

for i in students_df.columns:
    students_df[i]=students_df[i].fillna(0)
print(students_df)
```
